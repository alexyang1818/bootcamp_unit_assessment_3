```python
l_stops_df = pd.read_csv('CTA_list_of_L_stops.csv')
station_bools = l_stops_df[['MAP_ID','ADA','RED','BLUE','G','BRN','P','Pexp','Y','Pnk','O']].groupby('MAP_ID').any()
l_stations_df = l_stops_df[['MAP_ID','STATION_NAME','STATION_DESCRIPTIVE_NAME','Location']] \
    .merge(station_bools, how='left', left_on='MAP_ID', right_index=True).drop_duplicates()
l_stations_df.head()
```

```python
import re
stations = l_stations_df['Location'].str.split(r', ', expand=True).replace(r'\(|\)', '', regex=True)
stations_lat = pd.to_numeric(stations[0])
stations_lng = pd.to_numeric(stations[1])

l_stations_df['latitude'] = stations_lat
l_stations_df['longitude'] = stations_lng

l_stations_df.drop('Location', axis=1, inplace=True)
l_stations_df
```

```python
ridership_df = pd.read_csv('CTA_ridership_daily_totals.csv')
ridership_df
```

```python
import sqlalchemy
from sqlalchemy import create_engine
engine = create_engine(f"postgresql://postgres:xxxx@localhost:5432/cta_db")
l_stations_df.to_sql(name='stations', con=engine, index=False, if_exists='replace')
ridership_df.to_sql(name='ridership', con=engine, index=False, if_exists='replace')
```

```python
len(ridership_df['station_id'].unique())
```

```python
rider_station_df = ridership_df.merge(l_stations_df, how='left', left_on='station_id', right_on='MAP_ID')
rider_station_df
```

```python
rider_south_station_df = rider_station_df[rider_station_df['latitude']<=41.881]
rider_south_station_df['rides'].sum()
```
